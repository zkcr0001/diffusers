```python
# 获取图片长和宽
item = os.path.join(img_path, "image_{:04d}.png".format(1))
# image = imageio.imread(item)
image = Image.open(item)
image = np.array(image)
hed_edge = image
print(hed_edge.shape)
print(np.unique(hed_edge))
height = image.shape[0]
width = image.shape[1]
# 获取fps
fps = 30
plt.imshow(hed_edge)
image = Image.fromarray(hed_edge)
```

```python
video = cv2.VideoWriter(out_path, cv2.VideoWriter_fourcc('m', 'p', '4', 'v'), fps, (width, height))  # 创建视频流对象
```

```python
for i in range(file_num):
    item = os.path.join(img_path, "image_{:04d}.png".format(i))
    if os.path.exists(item):
        print("read",item)
        video.write(cv2.imread(item))  # 把图片写进视频
video.release()  # 释放
```
